```python
import pandas as pd

# define series 
data = pd.Series([1,1,1,2,3,3,3,3,4,4,5])

# Getting the frequencies of each value

data.value_counts() # It's reading like 1 is 3 times in the series
#data.value_counts(sort=False) # sort = False means it wont sort the data value by frequency
```

```python
# Creating data 

'''
df = dataframe 
'''

df = pd.DataFrame({'Grade':['A','A','A','B','B', 'B', 'B', 'C', 'D', 'D'],
                  'Age': [18, 18, 18, 19, 19, 20, 18, 18, 19, 19],
                  'Gender': ['M','M', 'F', 'F', 'F', 'M', 'M', 'F', 'M', 'F']})

df # to view the data
```

```python
'''
The way to interpret this is
--> 3 students recieved an 'A' in the class
--> 4 students recieved a 'B' in the class
'''


# Now let's look for the each latter grade
pd.crosstab(index=df['Grade'],columns='counts')
```

```python
'''
The way is to interpret this is -
--> 5 students are 18 years old
--> 4 students are 19 years old
--> 1 student is 20 years old
'''



# If we want to check age and then count it

pd.crosstab(index=df['Age'],columns='counts')
```

```python
'''
The way to interpret this is --
--> 50% of the students are 18 years old
--> 40% of the students are 19 years old
'''

# If we want to find out the relative frequncy table then

# Defining the crosstab
tab = pd.crosstab(index=df['Age'],columns='count')

# Lets get the proportion
tab/tab.sum() * 100
```

```python
'''
The way to interpret this is -
--> There are 3 students who are 18 years old and recieved an 'A' 
in the class
'''




# Now we can look how we can see two way frequency that means we are going to 
# see how we can see grade and age frequency at the same time

pd.crosstab(index=df['Age'], columns=df['Grade'])
```
