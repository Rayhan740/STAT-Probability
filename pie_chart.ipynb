```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ['English', 'Physics', 'Bangla', 'Arabic']
myexplode = [0,0.2,0,0] # Separating a particular data

plt.pie(y, labels = mylabels, startangle=0, explode = myexplode, shadow=True) # Creating a pie chart with start angle 0 and giving it a shadow
plt.legend(title='Subjects:') # Adding legend to the pie chart
plt.show()
```
